```python
import pandas as pd
import ast
```

# Data Cleaning for Marvel Data

```python
with open("marvel.txt") as f:
    string = f.read()
    f.close()
data_as_dict = ast.literal_eval(string)
pairings = data_as_dict["links"]

marvel_data = pd.DataFrame(pairings)
```

```python
source = []
target = []

for char in marvel_data.source:
    name = char.split("(", 1)
    source.append(name[0])
    
for char in marvel_data.target:
    name = char.split("(", 1)
    target.append(name[0]) 
    
dict_ = {"source": source, "target": target}
```

```python
marvel_data = pd.DataFrame(dict_)
```

```python
marvel_data.to_csv("marvel.csv")
```
